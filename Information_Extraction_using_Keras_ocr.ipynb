```python
import keras_ocr
import matplotlib.pyplot as plt
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
images = [
    keras_ocr.tools.read(img) for img in ['/content/images.png'
    ]
]
```

```python
len(images)
```

```python

plt.figure(figsize = (10,20))
plt.imshow(images[0])
```

```python
# Get Predictions

prediction_groups = pipeline.recognize(images)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(nrows=len(images), figsize=(10, 15))

# Flatten the axs array
axs_flat = np.ravel(axs)

for i, (ax, image, predictions) in enumerate(zip(axs_flat, images, prediction_groups)):
    keras_ocr.tools.drawAnnotations(image=image,
                                    predictions=predictions,
                                    ax=ax)
    ax.set_title(f"Image {i+1}")  # Set a title for each subplot if needed

plt.show()
```
